```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.petroldieselprice.com/petrol-price-previous-historical-trend-chart-in-Chennai/Tamil-Nadu'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
soup
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
table
```

```python
header = table.find_all('th')
header
```

```python
column_header = [i.text.strip() for i in header]
column_header
```

```python
df = pd.DataFrame(columns=column_header)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row = [data.text.strip() for data in row_data]
   
    
    length = len(df)
    df.loc[length] = individual_row
```

```python
df
```

```python
df.to_csv(r'F:\\My Project\\Petrol_price.csv',index = False)
```
