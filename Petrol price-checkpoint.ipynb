```python


# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="DATE", y= ' PETROL PRICE / Lit ', marker='o',color = '#23C9FF')
plt.title('Petrol Price Over Time')
plt.xlabel('Date')
plt.ylabel('Petrol Price / Lit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Calculate average petrol price for each month
monthly_avg_prices = df.groupby('MONTH')[' PETROL PRICE / Lit '].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_avg_prices, x='MONTH', y=' PETROL PRICE / Lit ',marker ='o',color ='#23C9FF')
plt.title('Average Petrol Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Petrol Price / Lit')
plt.tight_layout()
plt.show()
```

```python
# Calculate average petrol price for each year
yearly_avg_prices = df.groupby('YEAR')[' PETROL PRICE / Lit '].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_avg_prices, x='YEAR', y=' PETROL PRICE / Lit ', marker='o',color ='#23C9FF')
plt.title('Yearly Average Petrol Price')
plt.xlabel('Year')
plt.ylabel('Average Petrol Price / Lit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(df[' PETROL PRICE / Lit '], bins=20, color='#23C9FF', edgecolor='black')
plt.title('Distribution of Petrol Prices')
plt.xlabel('Petrol Price / Lit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plotting KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df[' PETROL PRICE / Lit '], color='#23C9FF', fill=True)
plt.title('Kernel Density Estimation (KDE) Plot of Petrol Prices')
plt.xlabel('Petrol Price / Lit')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Plotting box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df[' PETROL PRICE / Lit '], color='#23C9FF')
plt.title('Box Plot of Petrol Prices')
plt.xlabel('Petrol Price / Lit')
plt.grid(True)
plt.show()
```

```python
# Calculate monthly percentage change in petrol prices
df['MONTHLY_PERCENT_CHANGE'] = df[' PETROL PRICE / Lit '].pct_change() * 100

# Calculate yearly percentage change in petrol prices
df['YEARLY_PERCENT_CHANGE'] = df[' PETROL PRICE / Lit '].pct_change(12) * 100

# Plotting monthly percentage change
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='DATE', y='MONTHLY_PERCENT_CHANGE', marker='o',color ='#23C9FF')
plt.title('Monthly Percentage Change in Petrol Prices')
plt.xlabel('Date')
plt.ylabel('Monthly Percentage Change')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plotting yearly percentage change
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='DATE', y='YEARLY_PERCENT_CHANGE', marker='o',color='#23C9FF')
plt.title('Yearly Percentage Change in Petrol Prices')
plt.xlabel('Date')
plt.ylabel('Yearly Percentage Change')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
